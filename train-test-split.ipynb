```python
import pandas as pd 
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
ROOT = '/gdrive/MyDrive/kuliah/Tugas Akhir/Eksperimen'
data = pd.read_csv(f"{ROOT}/nested/TA_dataset_raw_labeled_nested_4th.csv")
data = data[data['juz'] <= 6]
data = data.fillna('O')

data.rename({
    'word_list' : 'Kata',
    'label_1st_pass' : 'Label',
    'pos_list' : 'POS',
    'Unnamed: 5' : 'nested'
}, inplace=True, axis=1)
data = data[data['Kata'] != '--']
data
```

```python
def preprocess(data) : 

    sentences = data.groupby(['juz', 'no_ayat'])['Kata'].apply(list).values

    # tag = data.groupby(['juz', 'no_ayat'])['nested'].apply(list).values
    tag = data.groupby(['juz', 'no_ayat'])['joint_tag'].apply(list).values

    pos = data.groupby(['juz', 'no_ayat'])['POS'].apply(list).values

    return sentences, tag, pos
```

```python
sentences, tag, pos = preprocess(data)
```

```python
data2 = pd.read_csv(f"{ROOT}/nested/TA_dataset_raw_labeled_nested_4th.csv")
data2 = data2[data2['juz'] <= 6]
data2 = data2[['juz', 'no_ayat']]
data2 = data2.groupby(['juz', 'no_ayat']).head(1)
data2['sent'] = sentences
data2['tag'] = tag
data2['pos'] = pos
data2['sentence #'] = np.arange(len(data2)) + 1

data2
```

```python
# data2.to_csv(f'{ROOT}/inference/TA_dataset_raw_labeled.csv', index=False)
```

```python
(
    train_data,
    test_data, 
) = train_test_split(
    data2,
    test_size=0.2,
    random_state=43
)
```

```python
train_data.to_csv(f"{ROOT}/nested/train_data.csv", index=False)
test_data.to_csv(f"{ROOT}/nested/test_data.csv", index=False)
```
