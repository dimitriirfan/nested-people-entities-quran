```python
import pandas as pd
from ast import literal_eval
```

```python
ROOT = "/gdrive/MyDrive/kuliah/Tugas Akhir/Eksperimen"
```

```python
train_data = pd.read_csv(f"{ROOT}/nested/train_data.csv")
test_data = pd.read_csv(f"{ROOT}/nested/test_data.csv")

train_data['sent'] = train_data['sent'].apply(literal_eval)
train_data['tag'] = train_data['tag'].apply(literal_eval)
train_data['pos'] = train_data['pos'].apply(literal_eval)


test_data['sent'] = test_data['sent'].apply(literal_eval)
test_data['tag'] = test_data['tag'].apply(literal_eval)
test_data['pos'] = test_data['pos'].apply(literal_eval)


train_data.head()
```

```python
train_data
```

```python
def explode_sent_tag(data) :

  data = train_data

  sent = data.explode('sent')['sent']
  tag = data.explode('tag')['tag']
  pos = data.explode('pos')['pos']

  new = data.explode('sent')
  new['sent'] = sent
  new['tag'] = tag
  new['pos'] = pos

  return new

def to_text(filepath, data) : 
  sentences = data['sentence #'].unique()
  col = ['sent', 'tag']

  with open(filepath, 'w') as f : 
    for sent in sentences : 
      df = data[data['sentence #'] == sent]
      for i, row in df.iterrows() : 
        f.writelines(f"{row['sent']} {row['tag']}\n")
      f.writelines("\n")
```

```python
new_train_data = explode_sent_tag(data=train_data)
new_test_data = explode_sent_tag(data=test_data)
```

```python
to_text(f'{ROOT}/nested/flair-train.txt', new_train_data)
to_text(f'{ROOT}/nested/flair-test.txt', new_test_data)
```
